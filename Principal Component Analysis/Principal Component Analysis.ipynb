# Principal Component Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("data/Breast_Cancer.csv")
dataset.head()
```

```python
dataset = dataset.drop(['Unnamed: 32','id'],axis = 1)
```

```python
dataset["diagnosis"].replace(to_replace = dict(M = 1, B = 0), inplace=True)
```

```python
x = dataset.iloc[:, 1:]
y = dataset["diagnosis"]

x, y = x.to_numpy(), y.to_numpy()
```

```python
class PrincipalComponentAnalysis:
    
    def __init__(self, n_components):
        self.n_components = n_components
        self.components = None
        self.mean = None
        
    def fit(self, X):
        # mean
        self.mean = np.mean(X, axis=0)
        X = X - self.mean
        
        # covariance
        cov = np.cov(X.T) # x(rows=samples, columns=featuers) = [[1, 2, 3], [4, 5, 6]] and covariance function tak values like (columns=samples, rows=featuers) x = x.T
       
        # eignvalues, eigenvectors
        eigenvalues, eigenvectors = np.linalg.eig(cov)
        
        # sort eigenvectors
        eigenvectors = eigenvectors.T
        idxs = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idxs]
        eigenvectors = eigenvectors[idxs]
        
        # sort first eigenvectors
        self.components = eigenvectors[0: self.n_components]
        
        
    def transform(self, X):
        # project data
        X = X - self.mean
        return np.dot(X, self.components.T)
```

```python
def plot(X_projected, y):
    x1 = X_projected[:, 0]
    x2 = X_projected[:, 1]

    plt.scatter(
        x1, x2, c=y, edgecolor="none", alpha=0.8, cmap=plt.cm.get_cmap("viridis", 3)
    )

    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.colorbar()
    plt.show()
```

```python
PCA = PrincipalComponentAnalysis(n_components=15)
PCA.fit(x)
X_projected = PCA.transform(x)

print("Shape of X:", x.shape, y.shape)
print("Shape of transformed X:", X_projected.shape)
print(f"Principal Component Analysis")

plot(X_projected, y)
```

```python
dataset = pd.read_csv("Data/diabetes_dataset.csv")

X = dataset.iloc[:, :-1]
Y = dataset["Outcome"]

X, Y = X.to_numpy(), Y.to_numpy()


pca = PrincipalComponentAnalysis(n_components=2)
pca.fit(X)
X_pro = pca.transform(X)

print("Shape of X:", X.shape, Y.shape)
print("Shape of transformed X:", X_pro.shape)
print(f"Principal Component Analysis")

plot(X_pro, Y)
```
